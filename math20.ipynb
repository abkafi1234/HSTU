```python
import pandas as pd
data = pd.read_csv("/content/50_Startups (1).csv")
```

```python
data
```

```python
# #missing value handelling
# 1. MCAR = Impute, Interpolation -> batil
# 2. MAR =
# 3. MNAR =
```

```python
data.isna().sum()
```

```python
data['Marketing Spend'] = data['Marketing Spend'].fillna(data['Marketing Spend'].mean())
```

```python
data
```

```python
data['State'].value_counts()
```

```python
data['State'].mode()[0]
```

```python
data['State'] = data['State'].fillna(data['State'].mode()[0])
```

```python
data
```

```python
data.info()
```

```python
data['R&D Spend'] = data['R&D Spend'].replace(0.0,data['R&D Spend'].mean())
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
data['State_t'] = lb.fit_transform(data['State'])
```

```python
data.head()
```

```python
lb.classes_
```

```python
data2 = data.drop(columns=['State'])
```

```python
data2.head()
```

```python
import seaborn as sns
sns.heatmap(data2.corr())
```

```python
data2 = data2[['R&D Spend', "Marketing Spend", 'Profit']]
```

```python
data2.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data2['Marketing Spend'], data2['Profit'])
```

```python
x = data2[['R&D Spend']]
y = data2['Profit']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2)
```

```python
xtrain,ytrain
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.score(xtest,ytest)
```

```python
lr.coef_ #m
```

```python
lr.intercept_ #c y=mx+c
```

```python
from sklearn.metrics import mean_squared_error
ypred = lr.predict(xtest)
error = mean_squared_error(ytest, ypred)**.5
```

```python
error
```

```python
predicted = lr.predict([[165349.20]])
```

```python
print(f'The predicted value will fall within {predicted - error} and {predicted + error}')
```
