```python
data2 = data[['R&D Spend', 'Profit']]
```

```python
data2.head()
```

```python
x = data2[['R&D Spend']]
y = data2['Profit']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain) #fit == train
lr.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_squared_error
ypred= lr.predict(xtest)
error = mean_squared_error(ytest, ypred)**.5
```

```python
predict = lr.predict([[162597.7000]])
```

```python
print(f'The range is {predict-error} and {predict+error}')
```
